```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
train_df = pd.read_csv('train (2).csv')
test_df = pd.read_csv('test.csv')
len_train_df = len(train_df)
len_test_df = len(test_df)
print(len_train_df)
```

```python
print(len_test_df)
```

```python
df = pd.concat([train_df, test_df], ignore_index = True)
df.head(10)
```

```python
df.info()
```

```python
df.isnull()
```

```python
 df.fillna(-999, inplace = True, axis = 1)
```

```python
df.drop(['Applicant_ID', 'form_field48', 'form_field49'], inplace = True, axis = 1)
df.info()
```

```python
df[['default_status', 'form_field47']].sample(10)
```

```python
df[['default_status', 'form_field47']]= df[['default_status', 'form_field47']].astype('category')
```

```python
df.form_field47 = df.form_field47.cat.codes
```

```python
df.default_status = df.default_status.cat.codes
```

```python
df[['default_status', 'form_field47']].sample(10)
```

```python
target_col = df.pop('default_status')
```

```python
df.head(10)
```

```python
target_col.value_counts()
```

```python
X = df.iloc[: len_train_df].copy()
print (X)
```

```python
y = target_col.iloc[: len_train_df]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
rf_clf = RandomForestClassifier(n_jobs = 1,
                               random_state = 2020,
                               n_estimators = 50).fit(X_train, y_train)
```

```python
yhat_positive = rf_clf.predict_proba(X_test)[:,1]
rf_clf.predict_proba(X_test)
```

```python
roc_auc_score(y_test, yhat_positive)
```
